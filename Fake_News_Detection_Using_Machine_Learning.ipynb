```python
import numpy as np
import pandas as pd
```

```python
true = pd.read_csv('/content/True.csv')
fake = pd.read_csv('/content/Fake.csv')
```

```python
true['label']= 1
fake['label']= 0
```

```python
news = pd.concat([fake,true],axis=0)
```

```python
news=news.drop(['title','subject','date'],axis=1)
```

```python
news = news.sample(frac=1)
```

```python
news.head()
```

```python
news.reset_index(inplace=True)
```

```python
news.head()
```

```python
news.drop(['index'],axis=1,inplace = True)
```

```python
news.head()
```

```python
import re
```

```python
def wordopt(text):
  text = text.lower() #lower convertion
   #remove urls
  text= re.sub(r'https?://\S+|www\.\S+','',text)
  #remove html tags
  text = re.sub(r'<.*?>','',text)
  #remove punctuation
  text = re.sub(r'[^\w\s]','',text)
  #remove digits
  text = re.sub(r'\d','',text)
  #remove new line characters
  text = re.sub(r'\n',' ',text)

  return text
```

```python
news['text'] = news['text'].apply(wordopt)
```

```python
news['text']
```

```python
x=news['text']
y=news['label']
```

```python
x
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.3)
```

```python
x_train.shape
```

```python
#convert categorical to numeric
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
vectorizer = TfidfVectorizer()
```

```python
xv_train = vectorizer.fit_transform(x_train)
```

```python
print(x_test[:5])  # Check the first 5 elements of x_test
```

```python
xv_test = vectorizer.transform(x_test)
```

```python
xv_test
```

```python
from sklearn.linear_model import LogisticRegression
```

LR = LogisticRegression()

```python
LR.fit(xv_train,y_train)
```

```python
pred = LR.predict(xv_test)
```

```python
LR.score(xv_test,y_test)
```

```python
from sklearn.metrics import classification_report
```

```python
print(classification_report(y_test,pred))
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
DTC = DecisionTreeClassifier()
```

```python
pred_dtc = DTC.predict(xv_test)
```

```python
DTC.score(xv_test,y_test)
```

```python
print(classification_report(y_test,pred_dtc))
```

```python
def output_label(n):
  if n==0:
    return "It is fake news"
  elif n==1:
    return "It is a Genuine news"
```

```python
def manual_testing(news):
    testing_news = {"text": [news]}  # Corrected syntax for defining dictionary
    new_def_test = pd.DataFrame(testing_news)
    new_def_test["text"] = new_def_test["text"].apply(wordopt)
    new_x_test = new_def_test["text"]
    new_xv_test = vectorizer.transform(new_x_test)  # Assuming 'vectorization' is your vectorizer object
    pred_lr = LR.predict(new_xv_test)
    # pred_dtc = dtc.predict(new_xv_test)
    pred_gbc = gbc.predict(new_xv_test)
    pred_rfc = rfc.predict(new_xv_test)

    return "\n\nLR Prediction: {} \nGBC Prediction: {} \nRFC Prediction: {}".format(
        output_label(pred_lr[0]), output_label(pred_gbc[0]), output_label(pred_rfc[0])
    )
```

```python
news_article = str(input())
```

```python
manual_testing(news_article)
```
